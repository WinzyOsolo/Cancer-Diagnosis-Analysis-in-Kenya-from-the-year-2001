# **4. Targeted Data Analysis**
In this section I will analyse the data to answer the key questions and purpose of this data analysis.

####a) Which are the top ten most prevalent type of Cancer in Kenya

We shall count and rank the data to establish which are the most prevalent cancers in Kenya

```python
# Rank the incidence of types of cancer in descending order
df['TYPE_RANK'] = df['CASES'].rank(ascending=False)

# Sort the DataFrame by rank in ascending order
df.sort_values(by='TYPE_RANK', ascending=True, inplace=True)


# Display the top 10 types of cancer
top_10_types = df[['TYPE_RANK','TYPE', 'CASES']].head(10)
print(top_10_types.head(10))
```

```python
# Visual representation of the data on top 10 types in the country
# Plotting the bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_10_types['TYPE'], top_10_types['CASES'], color='red')

# labels and title
plt.xlabel('Type of Cancer')
plt.ylabel('Number of Cases')
plt.title('Top 10 Types of Cancer by Number of Cases in Kenya')

# Rotating the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.show()
```

The top ten types of cancers are as above. This can help various research institutions to create targeted medicine and solutions to prepare for any incuidences if these cancers.

## b) Regional Analysis of the top five most prevalent cancers in Kenya

This is an analysis of the top five most prevalent cancers according to their regional distribution

```python
#Filter the data Frame to include only the top five cancers
specific_cancers = top_10_types['TYPE'].tolist()[:5]
specific_regions = ['NAIROBI', 'MOMBASA', 'NYANZA', 'CENTRAL', 'RIFT VALLEY','CENTRAL', 'EASTERN', 'NORTH_EASTERN' ]

filtered_df = df[df['TYPE'].isin(specific_cancers) & df[specific_regions].notnull().any(axis=1)]

#Group by cancer type and region, and sum the cases for each combination
cases_by_cancer_and_region = filtered_df.groupby('TYPE')[specific_regions].sum().reset_index()

# Display the DataFrame
print(cases_by_cancer_and_region)
```

There is a positive corelation between the two variables. Since the correlation coefficient is close to 1, it indicates that as the number of cases of types of cancers increases, the number of deaths also tends to increase. This does not necessarily imply that one causes the other.
